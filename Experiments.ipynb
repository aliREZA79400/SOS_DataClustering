```python
from mealpy.utils import io

import os
from utils import datasets

import numpy as np
```

```python
dataset , targ = datasets.retu_dataset("Iris_Dataset")
```

```python
def find_pkls(folder_path):
    targets = []
    for root, dirs, files in os.walk(folder_path):
        for name in files:
            
            if name.endswith((".pkl")) and name[8:12] == "Iris":
                
                file_path = os.path.join(root, name)

                model = io.load_model(file_path)
                
                targets.append((model.g_best.target.fitness,name[-5:-4]))
    return targets

```

```python
def find_pkls_new(folder_path,dataset_name):
    targets = []
    for root, dirs, files in os.walk(folder_path):
        for name in files:
            
            if name.endswith((".pkl")) and (dataset_name in name) :
                
                file_path = os.path.join(root, name)

                model = io.load_model(file_path)
                
                targets.append((model.g_best.target.fitness,name[-5:-4],file_path))
    return sorted(targets, key=lambda x: x[0])

```
