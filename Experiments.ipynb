# Collector

```python
from typing import  List , AnyStr , Tuple
from collections import  namedtuple
import os
from mealpy.bio_based.mySOS import mySOS
```

```python
model_results = namedtuple("model_results", ["algo_name","dataset_name","target", "run_number", "model_path","run_time"])
```

```python
def find_pkls_new(algorithm_path,dataset_name):
    targets = []
    for root, dirs, files in os.walk(algorithm_path):
        algo_name = os.path.basename(os.path.normpath(algorithm_path))
        for name in files:
            
            if name.endswith((".pkl")) and (dataset_name in name) :
                
                file_path = os.path.join(root, name)

                model = io.load_model(file_path)
                time = np.sum(model.history.list_epoch_time)
                targets.append(model_results(algo_name ,
                                             dataset_name, 
                                             model.g_best.target.fitness,
                                             name[-5:-4],file_path,
                                             time))
                                             
    return sorted(targets, key=lambda x: x[2])
```

```python
def collector_table3(model_paths :Tuple,
                     dataset_names:Tuple[AnyStr]):
    all_results = []
    for algo in model_paths:
        for dataset in dataset_names:
            all_results.append(find_pkls_new(algo,dataset))

    return all_results            
```

```python
model_paths = [ "./SOS",
                "./PSO",
                 "./ABC",
                 "./CS",
                 "./DE",
                 "./FPA",
                 "./MVO"]

dataset_names = ["Iris_Dataset" , "Breast_Cancer" , "Balance_Scale" , "Seeds" , "Statlog" , "Contraceptive_Method_Choice" , "Haberman_s_Survival", "Wine"]
```

```python
table_show = namedtuple("table3",["Algorithm", "Dataset", "Best", "Worst", "Mean", "Std","Average_Time"])

table_show_list = []

for algo_dataset in all_results:

    table_show_list.append(table_show(
        algo_dataset[0][0], # algorithm name same for all
        algo_dataset[0][1], # dataset name same for all
        algo_dataset[0][2], # best result 0 because is sorted
        algo_dataset[4][2], # worst index 4 because is sorted
        
        round(
            np.mean([algo_dataset[i][2] for i in range(len(algo_dataset))]),4),
        round(
            np.std([algo_dataset[i][2] for i in range(len(algo_dataset))]),4),
        round(
            np.mean([algo_dataset[i][5] for i in range(len(algo_dataset))]),4),
            )
    )
```
